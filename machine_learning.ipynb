4. What are outliers in a dataset, and list three common techniques for
handling them.

   .  Outliers are data points that significantly differ from other observations in a dataset.
Outliers are extreme values that lie an abnormal distance from other values in a random sample from a population.
Common techniques for handling outliers include:
Removal/Deletion: Removing the outlier data points from the dataset.
Imputation/Transformation: Replacing the outlier values with a more representative value (e.g., mean, median, or using a specific algorithm).
Capping/Flooring (Winsorization): Limiting the extreme values to a certain maximum or minimum threshold.

9. : Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,
20] using sklearn.preprocessing.MinMaxScaler. Print the scaled array.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# List of numbers
data = np.array([2, 5, 10, 15, 20]).reshape(-1, 1)

# Implement Min-Max scaling
scaler = MinMaxScaler()
scaled_array = scaler.fit_transform(data)

# Print the scaled array
print(scaled_array)
```
